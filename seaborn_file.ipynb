```python
import pandas as pd
import seaborn as sb
from matplotlib import pyplot as plt
df=sb.load_dataset('tips')

print(df.to_string())

g=sb.FacetGrid(df,col="time")
g.map(plt.hist,"tip")
plt.show()
```

```python
import pandas as pd
import seaborn as sb
from matplotlib import pyplot as plt
df=sb.load_dataset('tips')
g=sb.FacetGrid(df,col="time",hue="smoker")
g.map(plt.scatter,"total_bill","tip")
plt.show()
```

```python
import pandas as pd
import seaborn as sb
from matplotlib import pyplot as plt
df=sb.load_dataset('tips')
sb.barplot(x="sex",y="tip", hue="time",data=df)
plt.show()
```

```python
import pandas as pd
import seaborn as sb
from matplotlib import pyplot as plt
df=sb.load_dataset('tips')
sb.regplot(x="total_bill",y="tip",data=df)
plt.show()
```

```python
import pandas as pd
import seaborn as sb
from matplotlib import pyplot as plt
df=sb.load_dataset('tips')
sb.pairplot(df)
plt.show()
```

```python
import seaborn as sns
from matplotlib import pyplot as plt

df=sb.load_dataset('tips')
selected_vars=['total_bill','tip']
sns.pairplot(df,vars=selected_vars)
plt.show()
```

```python
import seaborn
from matplotlib import pyplot as plt
df=sb.load_dataset('tips')
seaborn.pairplot(df,hue='size')
plt.show()
```

```python
import seaborn as sns
from matplotlib import pyplot as plt

df=sb.load_dataset('tips')
seaborn.pairplot(df,hue='size',palette="husl")
plt.show()
```

```python
import seaborn as sns
from matplotlib import pyplot as plt

df=sb.load_dataset('tips')
seaborn.pairplot(df,kind='reg')
plt.show()
```
